```python
from ultralytics import YOLO

# 1. Load mô hình đã huấn luyện (đường dẫn đến file best.pt)
model = YOLO(r'D:\Project\segment car with yolo\train\train\weights\best.pt') 

# 2. Chạy dự đoán trên một hình ảnh
# conf: ngưỡng tin cậy (0.25 là mặc định)
# save: lưu kết quả vào thư mục 'runs/segment/predict'
results = model.predict(source='', conf=0.5, save=True)

# 3. Hiển thị kết quả (tùy chọn)
for result in results:
    result.show()  # Mở cửa sổ hiển thị hình ảnh đã vẽ mask
```

```python
import cv2
from ultralytics import YOLO

model = YOLO(r'D:\Project\segment car with yolo\train\train\weights\best.pt')
cap = cv2.VideoCapture('')

frame_count = 0
stride = 4

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    if frame_count % stride == 0:
        results = model.track(frame, persist=True, tracker='botsort.yaml', conf=0.5, retina_masks=True, iou=0.5)
        annotated_frame = results[0].plot(boxes=True, labels=True)
        cv2.imshow("YOLO", annotated_frame)
    
    frame_count += 1
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
